```python
import pandas as pd
import csv
```

```python
# load user data
users_path = '../../city_filter/users_Indianapolis.csv'
users_df = pd.read_csv(users_path, sep=',')
users_df
```

```python
business_path = '../business_features_engineering/businesses_Indianapolis_features.csv'
business_df = pd.read_csv(business_path, encoding='utf-8')
business_df
```

```python
reviews_path = "../../sentiment_analysis/reviews_Indianapolis_analyzed.csv"
reviews_df = pd.read_csv(reviews_path, encoding='utf-8')
reviews_df.head()
```

```python
# Step 1: 重命名用户数据中的投票字段，防止冲突
users_df_renamed = users_df.rename(columns={
    'useful': 'user_useful',
    'funny': 'user_funny',
    'cool': 'user_cool'
})

# Step 2: 合并评论与用户信息
merged = reviews_df.merge(users_df_renamed, on='user_id', how='left')

# Step 3: 处理账户年龄
merged['yelping_since'] = pd.to_datetime(merged['yelping_since'], errors='coerce')
merged['account_age_days'] = (pd.Timestamp.now() - merged['yelping_since']).dt.days
merged['elite_flag'] = merged['elite'].apply(lambda x: 0 if pd.isna(x) or x == '' else 1)

# Step 4: compliment 求和
compliment_cols = [col for col in users_df.columns if col.startswith('compliment_')]
merged['compliment_sum'] = merged[compliment_cols].sum(axis=1)

# Step 5: 只保留 business_df 中的商家
merged = merged[merged['business_id'].isin(business_df['business_id'])]

# Step 6: 聚合用户特征到商家层面
business_user_feature_df = merged.groupby('business_id').agg({
    'review_count': 'mean',
    'user_useful': 'mean',
    'user_funny': 'mean',
    'user_cool': 'mean',
    'average_stars': 'mean',
    'fans': 'mean',
    'elite_flag': 'mean',
    'compliment_sum': 'mean',
    'account_age_days': 'mean'
}).reset_index()

# Step 7: 重命名特征列
business_user_feature_df = business_user_feature_df.rename(columns={
    'review_count': 'avg_user_review_count',
    'user_useful': 'avg_user_useful',
    'user_funny': 'avg_user_funny',
    'user_cool': 'avg_user_cool',
    'average_stars': 'avg_user_avg_stars',
    'fans': 'avg_user_fans',
    'elite_flag': 'elite_user_ratio',
    'compliment_sum': 'avg_user_compliments',
    'account_age_days': 'avg_user_account_age_days'
})

business_user_feature_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 设置画图风格
sns.set(style="whitegrid")
plt.figure(figsize=(15, 10))

# 所有要可视化的数值列
num_cols = [
    'avg_user_review_count',
    'avg_user_useful',
    'avg_user_funny',
    'avg_user_cool',
    'avg_user_avg_stars',
    'avg_user_fans',
    'elite_user_ratio',
    'avg_user_compliments',
    'avg_user_account_age_days'
]

# 1. 分布图（hist + kde）
for i, col in enumerate(num_cols):
    plt.subplot(3, 3, i+1)
    sns.histplot(data=business_user_feature_df, x=col, kde=True, bins=30)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()

# 2. 相关性热力图
plt.figure(figsize=(10, 8))
sns.heatmap(business_user_feature_df[num_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()

# 3. 箱型图查看离群值
plt.figure(figsize=(15, 10))
for i, col in enumerate(num_cols):
    plt.subplot(3, 3, i+1)
    sns.boxplot(y=business_user_feature_df[col])
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

```python
# 需要进行 log1p 变换的列
import numpy as np
from sklearn.preprocessing import StandardScaler
log_transform_cols = [
    'avg_user_review_count',
    'avg_user_useful',
    'avg_user_funny',
    'avg_user_cool',
    'avg_user_fans',
    'avg_user_compliments'
]

# 创建副本
business_user_feature_df_log_scaled = business_user_feature_df.copy()

# 1. 添加 log1p 列
for col in log_transform_cols:
    business_user_feature_df_log_scaled[f'{col}_log'] = np.log1p(business_user_feature_df_log_scaled[col])

# 2. 标准化 log 列
scaler = StandardScaler()
log_scaled_cols = [f'{col}_log' for col in log_transform_cols]
business_user_feature_df_log_scaled[[f'{col}_scaled' for col in log_scaled_cols]] = scaler.fit_transform(
    business_user_feature_df_log_scaled[log_scaled_cols]
)

# 3. 构建最终输出列
final_columns = ['business_id',
                 'avg_user_avg_stars',
                 'elite_user_ratio',
                 'avg_user_account_age_days'] + [f'{col}_scaled' for col in log_scaled_cols]

business_user_feature_df_log_scaled = business_user_feature_df_log_scaled[final_columns]
business_user_feature_df_log_scaled
```

```python
## 重新做一遍刚刚的可视化
log_scaled_cols = [f'{col}_scaled' for col in log_scaled_cols]
# 设置画图风格
sns.set(style="whitegrid")
plt.figure(figsize=(15, 10))
# 1. 分布图（hist + kde）
for i, col in enumerate(log_scaled_cols):
    plt.subplot(3, 3, i+1)
    sns.histplot(data=business_user_feature_df_log_scaled, x=col, kde=True, bins=30)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()

# 更新列名
 

# 2. 相关性热力图
plt.figure(figsize=(10, 8))
sns.heatmap(business_user_feature_df_log_scaled[log_scaled_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
# 3. 箱型图查看离群值
plt.figure(figsize=(15, 10))
for i, col in enumerate(log_scaled_cols):
    plt.subplot(3, 3, i+1)
    sns.boxplot(y=business_user_feature_df_log_scaled[col])
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

```python
# 保存数据
output_path = 'businesses_user_Indianapolis_features.csv'
business_user_feature_df_log_scaled.to_csv(output_path, index=False)
```
