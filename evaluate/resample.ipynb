```python
import pandas as pd
```

```python
business_path = '../city_filter/businesses_Indianapolis.csv'
business_df = pd.read_csv(business_path, encoding='utf-8')
business_df
```

```python
# 随机生成从1 到 7546的整数，生成176个
import random
random.seed(0)
random_list = random.sample(range(1, 7546), 176)

business_df['is_choose'] = 0
for i in random_list:
    business_df.loc[business_df.index[i], 'is_choose'] = 1
business_df['is_choose'].value_counts()
```

```python
business_df.to_csv('businesses_Indianapolis_choose.csv', index=False, encoding='utf-8')
```
