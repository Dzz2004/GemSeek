```python
import pandas as pd
import numpy as np
```

```python
business_choose_now_path = 'businesses_Indianapolis_choose_now.csv'
business_choose_now = pd.read_csv(business_choose_now_path)

gem_now_path = 'gem_businesses_info_now.csv'
gem_now = pd.read_csv(gem_now_path)
```

```python
#business_choose_now 除去有缺失项的行
business_choose_now = business_choose_now.dropna()
```

```python
# 画出两个df的growth_rate_review的分布，画到一张图上
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.histplot(business_choose_now['growth_rate_review'], bins=100, color='blue', label='business_choose_now', kde=True)

plt.title('Distribution of growth_rate_review')
plt.xlabel('growth_rate_review')
plt.ylabel('Frequency')
# 标出频数最高的区间
plt.axvline(x=business_choose_now['growth_rate_review'].mean(), color='blue', linestyle='--', label='business_choose_now mean')
plt.axvline(x=business_choose_now['growth_rate_review'].median(), color='blue', linestyle=':', label='business_choose_now median')
plt.legend()
```

```python
# business_choose_now['growth_rate_review']的均值和中位数
mean_business_choose_now = business_choose_now['growth_rate_review'].mean()
median_business_choose_now = business_choose_now['growth_rate_review'].median()
print(f'business_choose_now mean: {mean_business_choose_now}')
print(f'business_choose_now median: {median_business_choose_now}')

# 划分为50个bins，给出频数最高的区间
hist, bin_edges = np.histogram(business_choose_now['growth_rate_review'], bins=50)
max_bin_index = np.argmax(hist)
max_bin_value = hist[max_bin_index]
max_bin_range = (bin_edges[max_bin_index], bin_edges[max_bin_index + 1])
print(f'Max bin range: {max_bin_range}')
```
